```python
# @markdown <br><center><img src='https://raw.githubusercontent.com/HenryZumaeta/DataProgrammingRepo/main/db/GDrive_to_GDrive_2023.png' height="86" alt="GD2GD-logo"/></center>
# @markdown  <center><h2><font size="5px" color="#FFBA00">Transfer</font> <font size="5px" color="#00AC47">from </font> <font size="5px" color="#2684FC">GDrive</font> <font size="5px" color="#EA4335">to</font> <font size="5px" color="#0066DA">GDrive</font></h2></center><br>

#@title `2. Drive Public Link Download`

# @markdown <br><left><h3>
# @markdown <font size="5px" color="#5D6368">Step 2 :</font>
# @markdown <font size="5px" color="#5D6368">Input</font>
# @markdown <font size="5px" color="#FFBA00">Source</font>
# @markdown <font size="5px" color="#00AC47">and</font>
# @markdown <font size="5px" color="#2684FC">Destination</font>
# @markdown <font size="5px" color="#EA4335">Path</font>
# @markdown </h3></left><br>

import subprocess

# Definir funciones
def run_shell_command(command):
    try:
        return subprocess.check_output(command, shell=True).decode().strip()
    except subprocess.CalledProcessError as e:
        print(f"Error al ejecutar el comando: {e}")
        return None

def count_files(path):
    return int(run_shell_command(f"find '{path}/' -type f | wc -l"))

def get_folder_size(path):
    return run_shell_command(f"du -sh '{path}'").split()[0]

# Definir rutas de origen y destino
INPUT_PATH_OF_GDRIVE = "/content/drive/MyDrive/Folder01" #@param {type:"string"}
OUTPUT_PATH_OF_GDRIVE = "/content/drive/MyDrive/Folder02" #@param {type:"string"}

# Contar archivos antes de copiar
num_files_ori = count_files(INPUT_PATH_OF_GDRIVE)

# Extraer el nombre de la última carpeta en la ruta de origen
folder_name = INPUT_PATH_OF_GDRIVE.rstrip('/').split('/')[-1]
output_path_with_folder = f"{OUTPUT_PATH_OF_GDRIVE}/{folder_name}"

# Mensaje de inicio de copia
print(f"Copiando {num_files_ori} archivos...", end='', flush=True)

# Realizar la copia
subprocess.run(["cp", "-a", INPUT_PATH_OF_GDRIVE + "/", output_path_with_folder])

# Borrar mensaje anterior
print("\r", end='')

# Contar archivos después de copiar y obtener tamaño de la carpeta
num_files_copied = count_files(output_path_with_folder)
folder_size = get_folder_size(output_path_with_folder)

# Imprimir resultados
print(f"Número de archivos originales: {num_files_ori}")
print(f"Número de archivos copiados: {num_files_copied}")
print(f"Peso total de la carpeta copiada: {folder_size}\n")
print(f"Guardado en la ruta:\n{output_path_with_folder}")
```

```python
# @markdown <br><center><img src='https://raw.githubusercontent.com/HenryZumaeta/DataProgrammingRepo/refs/heads/main/db/GDrive_Directory_Summary_SF.png' height="200" alt="DirSummary-logo"/></center>
# @markdown <center><h3>
# @markdown <font size="5px" color="#FFBA00">G</font><font size="5px" color="#00AC47">D</font><font size="5px" color="#2684FC">r</font><font size="5px" color="#EA4335">i</font><font size="5px" color="#00832D">v</font><font size="5px" color="#0066DA">e</font>
# @markdown <font size="5px" color="#5D6368"> Directory Summary</font>
# @markdown </h3></center><br>

#@title `3. File and Directory Count`

# @markdown <br><left><h3>
# @markdown <font size="5px" color="#5D6368">Define the </font>
# @markdown <font size="5px" color="#2684FC">folder</font>
# @markdown <font size="5px" color="#5D6368"> to </font>
# @markdown <font size="5px" color="#FFBA00">analyze</font>
# @markdown <font size="5px" color="#5D6368">:</font>
# @markdown </h3></left><br>

import subprocess

# ==========================
# Funciones auxiliares
# ==========================
def run_shell_command(command):
    try:
        return subprocess.check_output(command, shell=True).decode().strip()
    except subprocess.CalledProcessError as e:
        print(f"Error al ejecutar el comando: {e}")
        return None

def count_files(path):
    """Cuenta el número de archivos (excluye carpetas)."""
    return int(run_shell_command(f"find '{path}' -type f | wc -l"))

def count_folders(path):
    """Cuenta el número de carpetas (excluye la raíz)."""
    return int(run_shell_command(f"find '{path}' -mindepth 1 -type d | wc -l"))

def get_folder_size(path):
    """Obtiene el tamaño de la carpeta: legible y en bytes con dos decimales."""
    size_bytes_output = run_shell_command(f"du -sb '{path}'")

    if size_bytes_output is None:  # Si el comando falló
        print(f"⚠️ No se pudo obtener el tamaño de: {path}")
        return "0.00 B", "0"  # Valores por defecto para evitar errores

    try:
        size_bytes_str = size_bytes_output.split()[0]
        size_bytes = int(size_bytes_str)

        def convert_bytes(size):
            units = ['B', 'KB', 'MB', 'GB', 'TB']
            i = 0
            while size >= 1024 and i < len(units) - 1:
                size /= 1024.0
                i += 1
            return f"{size:.2f} {units[i]}"

        size_human = convert_bytes(size_bytes)
        return size_human, size_bytes_str

    except (IndexError, ValueError) as e:
        print(f"Error al procesar el tamaño: {e}")
        return "0.00 B", "0"  # Valores por defecto

def check_copy_status(num_origen, num_destino):
    """Verifica si los archivos copiados coinciden con los esperados."""
    if not isinstance(num_origen, int) or not isinstance(num_destino, int):
        return None

    if num_origen == num_destino:
        return "✅ Se copiaron todos los archivos correctamente."
    else:
        return (
            "❌ No se copiaron todos los archivos.\n"
            f"Esperados: {num_origen} | Copiados: {num_destino}"
        )

# ==========================
# Ruta de origen
# ==========================
INPUT_PATH_OF_GDRIVE = "/content/drive/MyDrive/Folder02"  #@param {type:"string"}

# ==========================
# Evaluación
# ==========================
num_files_ori2 = count_files(INPUT_PATH_OF_GDRIVE)
num_folders = count_folders(INPUT_PATH_OF_GDRIVE)
size_human, size_bytes = get_folder_size(INPUT_PATH_OF_GDRIVE)

# ==========================
# Resultados
# ==========================
print(f"Número de archivos: {num_files_ori2}")
print(f"Número de carpetas: {num_folders}")
print(f"Peso total de la carpeta: {size_human} ({size_bytes} bytes)\n")

# ==========================
# Chequeo de copia
# ==========================
try:
    check = check_copy_status(num_files_ori, num_files_ori2)
    if check:
        print(f"Chequeo:\n{check}")
except NameError:
    pass
```
